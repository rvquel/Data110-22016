```python
### 1. Setup and Styles
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print(plt.style.available)
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use('ggplot')
```

```python
### 2. Create a plot of the Quadratic Function
def f(x):
  return x**2 - 2*x + 1
x_values = np.linspace(-2, 4, 400)
y_values = f(x_values)
plt.plot(x_values, y_values, label="f(x) = x^2 - 2x + 1")
plt.title("Plot of the Quadratic Function f(x) = x^2 - 2x + 1")
plt.xlabel("x")
plt.ylabel("f(x)")
```

```python
### 3. Multiple Lines

x1 = np.linspace(-2,2,100)
x2 = np.linspace(0.01, 3, 100)

f = x1**3
g = np.exp(x1)
h = np.log(x2)

# Plot the functions with different colors
plt.plot(x1, f, label="f(x) = x^3", color='red')
plt.plot(x1, g, label="g(x) = exp(x)", color='blue')
plt.plot(x2 , h ,label="h(x) = log(x)", color='green')

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")

# Add grid and legend
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

# Show the plot
plt.show()
```

```python
### 3. Multiple Lines & 4. Styling Lines
def f(x):
    return x**2
def g(x):
    return x**3
def h(x):
    return x**4
def i(x):
    return x**5

# Generate x values
x_values = np.linspace(-2, 2, 400)

# Plot each function with a different color and line style
plt.plot(x_values, f(x_values), 'b-', label="f(x) = x^2")       # Blue solid line
plt.plot(x_values, g(x_values), 'g--', label="g(x) = x^3")      # Green dashed line
plt.plot(x_values, h(x_values), color='#FF5733', linestyle=':', label="h(x) = x^4")  # Custom hex color with dotted line
plt.plot(x_values, i(x_values), color=(0.5, 0.1, 0.8), linestyle='-.', label="i(x) = x^5")  # RGB tuple with dash-dot line

# Add labels and title
plt.title("Plot of functions with different line styles and colors")
plt.xlabel("x")
plt.ylabel("y")

# Add grid and legend
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

# Show the plot
plt.show()
```

```python
### Labels and Legends
def f(x):
    return x
def g(x):
    return x**(1/3)

# Generate x values for the range x >= 0
x1 = np.linspace(0, 10, 400)

# Compute y values
y_f = f(x1)
y_g = g(x1)

# Plot the functions
plt.plot(x1, y_f, label="f(x) = x", color='blue')
plt.plot(x1, y_g, label="g(x) = x^(1/3)", color='green')

# Add labels and title
plt.title("Plot of f(x) = x and g(x) = x^(1/3)")
plt.xlabel("x")
plt.ylabel("y")

# Add grid and legend
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()

# Show the plot
plt.show()
```

```python
### 8.Customizing Scatter Plots
x = np.linspace(0,4*np.pi,100)

#writing the function
y = x * np.sin(x)

#added labels and ploted the graph and scatter of points
plt.plot(x,y, label = 'line', color = 'blue', linewidth = '4')
plt.scatter(x,y, label = 'scatter points', color = 'red', edgecolor = 'black', s=100)
plt.xlabel('range of values')
plt.ylabel('plot of graphs')
plt.legend()
```

```python
### 9. Combined Plot
X = np.linspace(-2*np.pi,2*np.pi,100)
Y = X*np.cos(X)

#plot the scatter plot with intervals every pi/4
plt.plot(X,Y, label = 'Line of data', color = 'black', linestyle = ':')
x_scatter = np.arange(-2*np.pi, 2*np.pi + np.pi/4, np.pi/4)
y_scatter = x_scatter * np.cos(x_scatter)
plt.scatter(x_scatter,y_scatter, label = 'Scatter points', color = 'green')

#labels for the graph
plt.xlabel('range of values')
plt.ylabel('Plot of graphs')
plt.title('Graph of xcos(x)')
plt.legend()
```

```python
### 10. Custom Plot with Alternative Style
plt.style.use('ggplot')

# Define the functions
def f(x):
    return np.tan(x)

def g(x):
    # Handle the case where x = 0
    return np.sinc(x / np.pi)  # sinc(x) = sin(pi*x)/(pi*x)

# Generate x values for the range of interest
x_values = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
x_values_g = np.linspace(-10, 10, 1000)  # Larger range for sinc(x)

# Compute y values, with appropriate domain limitations for tan(x)
y_f = f(x_values)
y_g = g(x_values_g)

# Custom scatter points (key points to highlight)
scatter_x = [0, np.pi, -np.pi]
scatter_y = g(np.array(scatter_x))

# Plot the functions
plt.plot(x_values, y_f, label="f(x) = tan(x)", color='blue', linestyle='-', linewidth=1)
plt.plot(x_values_g, y_g, label="g(x) = sinc(x)", color='green', linestyle='--', linewidth=2)
plt.scatter(scatter_x, scatter_y, color='red', zorder=5, label="Key Points on g(x)")

# Add labels and title
plt.title("Plot of f(x) = tan(x) and g(x) = sinc(x)")
plt.xlabel("x")
plt.ylabel("y")

# Limit the axes to focus on key regions
plt.xlim(-2*np.pi, 2*np.pi)
plt.ylim(-10, 10)

# Add grid and legend
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
### 11. Style Comparison
#label x as the range of values
x = np.linspace(-2, 4, 400)
plt.style.use('default')
#labely as the quadratic equation
y = x**2-2*x+1

#graph the function from the range of values
plt.subplot(3,1,3)
plt.plot(x,y)
plt.ylabel('f(x)')
plt.xlabel('range of values')
#label x as the range of values
x = np.linspace(-2, 4, 400)

plt.style.use('seaborn-whitegrid')
#labely as the quadratic equation
y = x**2-2*x+1

#graph the function from the range of values
plt.subplot(3,1,1)
plt.plot(x,y)
plt.title("Quadratic function")
plt.ylabel('f(x)')
plt.xlabel('range of values')


#label x as the range of values
x = np.linspace(-2, 4, 400)
plt.style.use('ggplot')
#labely as the quadratic equation
y = x**2-2*x+1

#graph the function from the range of values
plt.subplot(3,1,2)
plt.plot(x,y)
plt.ylabel('f(x)')
plt.xlabel('range of values')

```

The first plot is the most suitable for analyzing data because it provides better visibility of the x and y values. The ggplot version has a harder to read color which makes it harder to interpret.The default one lacks legability as it is defined due to the lack of background.
