```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")  # Sets the style for the plots
```

```python
data = { 'Offense': ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery'],
    'Offenses per 100,000 people (2023)': [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5],
    'Change in Offenses per 100,000 People (2022-2023)': [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]}
```

```python
## Task 1: Basic Bar Chart for Crime Rates in 2023
df = pd.DataFrame(data)
colors = ['blue', 'darkred', 'green', 'yellow', 'hotpink', '#000000', '#9992BC']

plt.figure(figsize=(10, 6))
bars = plt.bar(df['Offense'], df['Offenses per 100,000 people (2023)'], color=colors)

plt.grid(axis='y', linestyle='--', alpha=0.8, color= 'black')
plt.xlabel('Crime Type')
plt.ylabel('Offense Rate (per 100,000 people)')
plt.title('Offense Rates in 2023')


plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
## Task 2: Change in Crime Rates (2022 vs. 2023) with Unique
# Each horizontal bar chart visualizes the percentage change in offense rates from 2022 to 2023 for each crime
```

```python
plt.figure(figsize=(10, 6))
bars = plt.barh(df['Offense'], df['Change in Offenses per 100,000 People (2022-2023)'], color=colors)

plt.grid(axis='x', linestyle='--', alpha=0.7)
for index, value in enumerate(df['Change in Offenses per 100,000 People (2022-2023)']):
    plt.text(value, index, f'{value}%', va='center')

plt.xlabel('Percentage Change in Offenses (per 100,000)')
plt.title('Percentage Change in Crime Rates from 2022 to 2023')
plt.ylabel('Offenses committed')

plt.grid(axis = 'y', linestyle = '--', color = 'black')

plt.tight_layout()
```

```python
## Why does this bar graph make more sense than Task 1's bar graph?
# Task #2 bar graph makes more sense than the Task #1, given you are able to analysis two data sets in two different years rather than just a snaphot of one year.
# By comparing data between two years you are able to draw conclusion of what progress (or regression) trend is being seen. Overall, this allows for the audiance to see a trend through percentages versus just seeing one year as an absolute number which can be misleading. In the case of this data set, we may believe Larceny is highest concern in crime rate from Task #2 but when comparing crime rates between 2023 and 2024, we can now believe in general Motor accidents have a higher concern in crime rate.
```

```python
## Task 2 (Alternate Version)
plt.figure(figsize=(10, 6))
bars = plt.barh(df['Offense'], df['Change in Offenses per 100,000 People (2022-2023)'], color=['darkgreen' if x < 0 else 'red' for x in df['Change in Offenses per 100,000 People (2022-2023)']])

plt.grid(axis='x', linestyle='--', alpha=0.7)
for index, value in enumerate(df['Change in Offenses per 100,000 People (2022-2023)']):
    plt.text(value, index, f'{value}%', va='center')

plt.xlabel('Percentage Change in Offenses (per 100,000)')
plt.title('Percentage Change in Crime Rates from 2022 to 2023')
plt.ylabel('Offenses committed')

plt.grid(axis = 'y', linestyle = ':', color = 'grey')

plt.tight_layout()
```
